```python
import ssaCore
import numpy as np
import matplotlib.pyplot as plt
import parser
%pylab inline
pylab.rcParams['figure.figsize'] = (8, 6)
```

```python
N = 200
L = 90 # Длинна гусеницы
K = N - L + 1 # Количество столбцов в траекторной матрице
t = np.arange(0,N)
```

```python
par = parser.par()
```

```python
trend = 0.08 * t
p1, p2 = 11, 8
periodic1 = 0.9 * np.sin(2*np.pi*t/p1)
periodic2 = 0.8 * np.cos(5*np.pi*(t+0.09)/p2)
```

```python
noise = np.random.normal(0, 0.5, N)

F = trend + periodic1 + periodic2 + noise
plt.plot(t, F, label="F", color = "black")
# plt.plot(t, trend, label="Trend", color = "red")
plt.title("Original time series")
plt.legend(prop={'size': 14})
plt.show()
```

```python
ssa = ssaCore.SSA1D(F, L)
ssa.getComponents()
print(ssa.components)
ssa.filterComponents()
print(ssa.components)
```

```python
tsReconstruct = ssa.reconstruct()
```
